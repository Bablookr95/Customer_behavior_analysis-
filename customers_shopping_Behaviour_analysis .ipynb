#### Data Cleaning

```python
data.isnull().sum()
```

```python
data['Purchase Amount (USD)'].fillna(data['Purchase Amount (USD)'].median(),inplace=True)
```

```python
data['Size'].fillna(data['Size'].mode()[0], inplace=True)
```

```python
data['Review Rating'].fillna(data['Review Rating'].mean(),inplace=True)
```

```python
data['Previous Purchases'].fillna(0).round(2)
```
